### Q1: What's the best month for sales?How much was earned that month?

### Add a "Sales" column

```python
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])
```

```python
df["Sales"] = df["Price Each"] * df["Quantity Ordered"]
```

```python
df
```

```python
results = df.pivot_table(values ="Sales", index = "Month", aggfunc = "sum" )
results
```

```python
import matplotlib.pyplot as plt
```

```python
results.plot(kind = "bar", y = "Sales")
plt.ylabel("Sales in USA Dollar")
plt.xlabel("Month")
```

### Question 2: What city had the highest sales?

```python
df["City"] = df["Purchase Address"].str.split(", ").str.get(1)
```

```python
df["city_code"] = df["Purchase Address"].str.split(", ").str.get(2).str.split(" ").str.get(0)
```

```python
df
```

```python
city_sales = df.pivot_table(values = "Sales", index = ["City", "city_code"], aggfunc = "sum")
city_sales
```

```python
city_sales.plot(kind = "bar", y = "Sales")
plt.ylabel("Income of City")
```

### Question 3: What time should we display advertisements to maximize the likelihood  of customer's buying products?

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Hour"] =  df["Order Date"].dt.hour
df["Minute"] = df["Order Date"].dt.minute
df.head()
```
